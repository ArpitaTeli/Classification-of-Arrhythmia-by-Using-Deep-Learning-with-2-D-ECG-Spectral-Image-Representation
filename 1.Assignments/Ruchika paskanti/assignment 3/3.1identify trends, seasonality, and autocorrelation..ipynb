```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df=pd.read_csv ('/content/Electric_Production.csv')
```

```python
# Preview the dataset
print(df.head())
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
```

```python
#visulize time series
plt.figure(figsize=(14, 7))
# Access the correct column name, in this case 'IPG2211A2N'
plt.plot(df['IPG2211A2N'], label='Electric Production')
plt.title('Electric Production Over Time')
plt.xlabel('Date')
plt.ylabel('Electric Production')
plt.legend()
plt.grid()
plt.show()
```

```python
decomposition = sm.tsa.seasonal_decompose(df['IPG2211A2N'], model='additive')  # Change 'Electric_Production' to 'IPG2211A2N'
decomposition.plot()
plt.show()

df['IPG2211A2N'].plot(figsize=(15, 6))  # Change 'Electric_Production' to 'IPG2211A2N'
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(211)
# Replace 'Electric_Production' with 'IPG2211A2N'
plot_acf(df['IPG2211A2N'].dropna(), lags=40, ax=plt.gca())
plt.title('Autocorrelation Function')

plt.subplot(212)
# Replace 'Electric_Production' with 'IPG2211A2N'
plot_pacf(df['IPG2211A2N'].dropna(), lags=40, ax=plt.gca())
plt.title('Partial Autocorrelation Function')

plt.tight_layout()
plt.show()
```
